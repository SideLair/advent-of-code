```python
with open('input.txt') as f:
    input = [tuple(line.split()) for line in f]
```

```python
input
```

```python
hor = 0
dep = 0

for command in input:
    if command[0] == 'forward':
        hor += int(command[1])
    elif command[0] == 'up':
        dep -= int(command[1])
    else:
        dep += int(command[1])

print(f'{hor=} * {dep=} = {hor * dep}')
```
